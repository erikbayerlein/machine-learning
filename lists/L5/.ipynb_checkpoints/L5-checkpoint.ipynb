```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import ConfusionMatrixDisplay, PrecisionRecallDisplay, RocCurveDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve
```

## Q1

### SVM

```python
data = pd.read_csv('../../datasets/californiabin.csv')
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=0)
```

```python
scaler = StandardScaler()
model = SVC(probability=True)
pipe = Pipeline(steps=[('scaler', scaler), ('svc', model)])
```

```python
params_SVC = [
    {'svc__kernel': ['rbf'], 
     'svc__gamma': 2 ** np.arange(-15.0, 3.0, 2),
     'svc__C': 2 ** np.arange(-5.0, 15.0, 2)}
]

svm_confusion_matrix = None
svm_roc_curve = None
svm_best_params = None
```

```python
cv = GridSearchCV(estimator=pipe, param_grid=params_SVC, cv=10, n_jobs=-1)

cv.fit(X_train, y_train)
model = cv.best_estimator_
svm_best_params = cv.best_params_
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
prediction_proba = model.predict_proba(X_test)
```

```python
acc = accuracy_score(y_test, prediction)
svm_confusion_matrix = confusion_matrix(y_test, prediction, normalize='true')
prec = precision_score(y_test, prediction)
rec = recall_score(y_test, prediction)
f1 = f1_score(y_test, prediction)

svm_metrics = np.array([acc, prec, rec, f1]).astype(float)
svm_roc_curve = roc_curve(y_test, prediction_proba[:, 1])
```

```python
print("="*60)
print("SVM - Métricas do Modelo")
print("="*60)
print(f"Acurácia:  {svm_metrics[0]:.4f}")
print(f"Precisão:  {svm_metrics[1]:.4f}")
print(f"Revocação: {svm_metrics[2]:.4f}")
print(f"F1-score:  {svm_metrics[3]:.4f}")
print(f"\nHiperparâmetros: {svm_best_params}")
print("="*60)

fig1, ax1 = plt.subplots(figsize=(8, 6))
cmdisplay = ConfusionMatrixDisplay(svm_confusion_matrix)
cmdisplay.plot(ax=ax1, cmap='YlOrRd')
ax1.set_title('SVM - Matriz de Confusão', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

fig2, ax2 = plt.subplots(1, 2, figsize=(14, 5))
prdisplay = PrecisionRecallDisplay.from_estimator(model, X_test, y_test, ax=ax2[0])
rocdisplay = RocCurveDisplay(fpr=svm_roc_curve[0], tpr=svm_roc_curve[1])
rocdisplay.plot(ax=ax2[1])

ax2[0].set_title('Curva Precision x Recall', fontsize=12, fontweight='bold')
ax2[1].set_title('Curva ROC', fontsize=12, fontweight='bold')
fig2.suptitle(f'SVM - Análise Completa', fontsize=14, fontweight='bold')
plt.tight_layout()
```

### Random Forest

```python
data = pd.read_csv('../../datasets/californiabin.csv')
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=0)
```

```python
params_rf = {
    'n_estimators' : [i for i in range(10, 210, 10)],
    'max_depth' : [4, 6, 8, 10, None]
}

rf_confusion_matrix = None
rf_roc_curve = None
rf_best_params = None
```

```python
model = RandomForestClassifier()
cv = GridSearchCV(estimator=model, param_grid=params_rf, cv=10, n_jobs=-1)
```

```python
cv.fit(X_train, y_train)
model = cv.best_estimator_
rf_best_params = cv.best_params_
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
prediction_proba = model.predict_proba(X_test)
```

```python
acc = accuracy_score(y_test, prediction)
rf_confusion_matrix = confusion_matrix(y_test, prediction, normalize='true')
prec = precision_score(y_test, prediction)
rec = recall_score(y_test, prediction)
f1 = f1_score(y_test, prediction)

rf_metrics = np.array([acc, rf_confusion_matrix[0][0], rf_confusion_matrix[1][1], prec, rec, f1]).astype(float)
rf_roc_curve = roc_curve(y_test, prediction_proba[:, 1])
```

```python
print("="*60)
print("Random Forest - Métricas do Modelo")
print("="*60)
print(f"Acurácia:  {rf_metrics[0]:.4f}")
print(f"Acc. [0]:  {rf_metrics[1]:.4f}")
print(f"Acc. [1]:  {rf_metrics[2]:.4f}")
print(f"Precisão:  {rf_metrics[3]:.4f}")
print(f"Revocação: {rf_metrics[4]:.4f}")
print(f"F1-score:  {rf_metrics[5]:.4f}")
print(f"\nHiperparâmetros: {rf_best_params}")
print("="*60)

fig1, ax1 = plt.subplots(figsize=(8, 6))
cmdisplay = ConfusionMatrixDisplay(rf_confusion_matrix)
cmdisplay.plot(ax=ax1, cmap='Blues')
ax1.set_title('Random Forest - Matriz de Confusão', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

fig2, ax2 = plt.subplots(1, 2, figsize=(14, 5))
prdisplay = PrecisionRecallDisplay.from_estimator(model, X_test, y_test, ax=ax2[0])
rocdisplay = RocCurveDisplay(fpr=rf_roc_curve[0], tpr=rf_roc_curve[1])
rocdisplay.plot(ax=ax2[1])

ax2[0].set_title('Curva Precision x Recall', fontsize=12, fontweight='bold')
ax2[1].set_title('Curva ROC', fontsize=12, fontweight='bold')
fig2.suptitle(f'Random Forest - Análise Completa', fontsize=14, fontweight='bold')
plt.tight_layout()
```
